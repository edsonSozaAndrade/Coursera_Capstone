```python
url="https://www.rug.nl/ggdc/docs/pwt91.xlsx"
gdpData=pd.read_excel(url, sheetname = 'Data')
```

```python
boliviaGdpData = pd.DataFrame(gdpData[gdpData.countrycode == 'BOL'].reset_index(drop=True))
boliviaGdpData.drop(labels=['countrycode', 'country', 'currency_unit', 
       'avh', 'hc', 'ccon', 'cda', 'cgdpe', 'cgdpo', 'cn', 'ck',
       'ctfp', 'cwtfp', 'rgdpna', 'rconna', 'rdana', 'rnna', 'rkna', 'rtfpna',
       'rwtfpna', 'labsh', 'irr', 'delta', 'xr', 'pl_con', 'pl_da', 'pl_gdpo',
       'i_cig', 'i_xm', 'i_xr', 'i_outlier', 'i_irr', 'cor_exp', 'statcap',
       'csh_c', 'csh_i', 'csh_g', 'csh_x', 'csh_m', 'csh_r', 'pl_c', 'pl_i',
       'pl_g', 'pl_x', 'pl_m', 'pl_n', 'pl_k'], axis=1,inplace = True)
```

```python
boliviaGdpData = boliviaGdpData.query('2009 <= year <= 2019')
boliviaGdpData = boliviaGdpData.append({
                        'year' : int(2018),
                        'rgdpe' : round(boliviaGdpData['rgdpe'].mean(),6),
                        'rgdpo' : round(boliviaGdpData['rgdpo'].mean(),6),
                        'pop' : round(boliviaGdpData['pop'].mean(),6),
                        'emp' : round(boliviaGdpData['emp'].mean(),6)
                    }, ignore_index=True)
boliviaGdpData = boliviaGdpData.append({
                        'year' : int(2019),
                        'rgdpe' : round(boliviaGdpData['rgdpe'].mean(),6),
                        'rgdpo' : round(boliviaGdpData['rgdpo'].mean(),6),
                        'pop' : round(boliviaGdpData['pop'].mean(),6),
                        'emp' : round(boliviaGdpData['emp'].mean(),6)
                    }, ignore_index=True)
```
